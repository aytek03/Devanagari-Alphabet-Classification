```python
num_pixels = X.shape[1]
num_classes = 602
img_width = 32
img_height = 32
img_depth = 1

#X_images = X.values.reshape(X.shape[0], img_width, img_height)

#for i in range(1, 9):
#    plt.subplot(240+i)
#    plt.imshow(X_images[i], cmap=plt.get_cmap('gray'))
#plt.show()
print(X.shape[0])
print(X.shape[1])
print(X.shape)
```

```python
# output in binary/one-hot encoded format
binencoder = LabelBinarizer()
Y = binencoder.fit_transform(Yd)
# data normalization
X /= 255

seed = 2
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=seed)
```

# 3. Training on Simple Neural Network

```python
# Training on Simple Neural Network
def nn_model():
    model = Sequential()
    model.add(Dense(num_pixels, input_dim=num_pixels, kernel_initializer='normal', activation='relu'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
nn = nn_model()


history_nn = nn.fit(X_train.values, y_train, validation_split=0.20, epochs=100, batch_size=400, verbose=2)
```

```python
scores_nn = nn.evaluate(X_test.values, y_test, verbose=0)
print('Accuracy: %.2f%%' % (scores_nn[1]*100))
```
